### Programowanie obiektowe

W poprzednim zadaniu dużym utrudnieniem była synchronizacja zestawów danych. Można to było zrobić na wiele sposobów. Mogliśmy zaangażować do tego słowniki, umieszczać dane na listach lub krotkach albo zwyczajnie zapisywać dane w jednej dużej liście i pilnować odpowiedniej organizacji indeksów. Naturalnym wydaje się organizowanie danych w **zamkniętych zestawach** i dopiero te zestawy w całości trzymać na wspólnej liście i wywoływać odpowiednie funkcje dla konkretnego zestawu. Właśnie na tej idei opiera się **programowanie obiektowe**.

**Obiekt** to podstawowe pojęcie programowaniu obiektowym i jest to pewna jednostka, która posiada własny zestaw zmiennych nazywanych **atrybutami** oraz własny zestaw funkcji nazywanych **metodami**, które są wykonywane na atrybutach tego obiektu. Obiekty są tworzone na podstawie ustalonego zestawu instrukcji, który nazywamy **klasą**. Tak samo jak jedna funkcja może zostać wywołana w programie wiele razy tak z jednej klasy można stworzyć wiele obiektów, są to tzw. **instancje** tej klasy. O klasach można myśleć jak o nowym typie danych, który sami konstruujemy.

Podstawowa składnia nowej klasy w Pythonie wygląda następująco:
```Python
class nazwa_klasy:
    def metoda1(self):
        self.atrybut1 = x
        self.atrybutN = y

        self.metodaN()

    def metodaN(self):
        suma = self.atrybut1 + self.atrybutN

        return suma
```

Dodatkowo, mamy możliwość tworzenia atrybutów i metod klasy. Oznacza to, że dana wartość i funkcjonalność będą współdzielone przez wszystkie instancje tej klasy. Składnia wóczas wygląda w taki sposób:
```Python
class nazwa_klasy:
    atrybut1 = x
    atrybut2 = y

    def metoda(self):
        self.atrybut1 = z

        return self.atrybut1 + nazwa_klasy.atrybut1
    
    def metoda_klasy():
        return nazwa_klasy.atrybut2
```

Warto zauważyć tutaj dwie rzeczy:
* metoda klasy nie przyjmuje parametru `self` - to ją odróżnia od metody instancji,
* w metodzie klasy nie wolno używać słowa kluczowego `self` ani odwoływać się do danych instancji.

W powyższych zadaniach kłopotliwe było to, że musieliśmy oddzielnie wywołać każdy atrybut naszego obiektu i przypisać do niego wartość. Dodatkowo, wewnątrz klasy musieliśmy operować na atrybutach, które de facto nie istniały. Jest to kłopotliwe. Można napisać metodę, którą będzie trzeba wywołać na obiekcie. Taka metoda przyjmowała by jako parametry dane jakie chcemy przypisać atrybutom obiektu i zawierała ewentualny dodatkowy kod do uruchomienia na świeżym obiekcie. Jeszcze fajniej gdyby taka metoda wywoływała się sama. 

Taką metodą jest tzw. konstruktor, czyli specjalna metoda, która jest wywoływana przez interpreter Pythona automatycznie podczas tworzenia obiektu. Parametry są do niej przekazywane w nawiasach podczas tworzenia obiektu, np. `obiekt = klasa(a, b)`. Najczęściej wykorzystuje się ją do inicjalizacji atrybutów obiektu i przypisywania do niego danych i sprawdzania poprawności tych danych. Konstruktor jest definiowany jako metoda o nazwie `__init__` i należy do całego grona tzw. metod magicznych: [https://docs.python.org/3/reference/datamodel.html](https://docs.python.org/3/reference/datamodel.html). 

Przykładowa klasa z konstruktorem:

```Python
class nowa_klasa:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

    def suma(self):
        return self.a + self.b + self.c
```

Dzięki takiej konstrukcji klasy możemy zadbać o to, aby obiekt na pewno posiadał odpowiednie dane przed wywołaniem metod, które ich używają. Zwiększa to również czytelność kodu i pozwala lepiej organizować klasę.

Dla zabezpieczenia przed niepoprawnym użyciem służą tzw. **dekoratory**. Konkretnie `@property` i `setter`. Dekorator `@property` pozwala *udekorować* metodę w taki sposób aby na zewnątrz była ona widziana jako atrybut, tzn. można jej używać bez nawiasów wywołujących. Metoda ta nie przyjmuje żadnych parametrów poza `self` i musi zwracać jakąś wartość. Reprezentuje akcję odczytania wartości atrybutu. Ten dekorator zawsze występuje wspólnie z dekoratorem `setter`, który służy do określenia co ma się wydarzyć kiedy do atrybutu przypisywana jest wartość. Przyjmuje parametry `self` i wartość przypisywaną do atrybutu. Niczego nie zwraca. Używamy ich w następujący sposób:

```Python
class klasa:
    def __init__(self, parametr: int):
        self.parametr = parametr

    @property
    def parametr(self):
        return self._parametr_
    @parametr.setter
    def parametr(self, wartosc):
        if isinstance(wartosc, int):
            self._parametr_ = wartosc
        else:
            raise AttributeError("Wartość parametru musi być liczbą całkowitą")
```

Wówczas takiego parametru można używać jak zwykłego atrybutu, ale Python dba o to aby miał on odpowiednią wartość.

Warto zwrócić tutaj uwagę na funkcję `isinstance`, która służy do sprawdzenia czy badany obiekt (w Pythonie wszystko jest obiektem) jest instancją podanej klasy. Drugą, wartą odnotowania nowością jest instrukcja `raise`. Służy ona do generowania wyjątków w programie. Dzięki temu nasz obiekt da znać użytkownikowi, że został użyty w niewłaściwy sposób.

#### Zadanie 7.
Zaktualizować klasę `Osoba` o odpowiednie mechanizmy zabezpieczeń. W przypadku próby przypisania danych, które nie spełniają założeń program powinien wypisać odpowiedni komunikat na ekran. Atrybuty `imie` i `nazwisko` mają być tekstem o długości od $3$ do $20$ znaków. Atrybut `numer_buta` ma być liczbą całkowitą z zakresu od $20$ do $50$. Podpowiedź: do sprawdzenia czy dane są odpowiedniego typu można użyć funkcji `isinstance()`, np.:
```Python
>>> a = 1
>>> isinstance(a, int)
True
>>> isinstance(a, str)
False
```

````python
class Osoba:
    def __init__(self, imie, nazwisko, numer_buta):
        # tutaj inicjalizacja atrybutów
        self.imie = imie
        self.nazwisko = nazwisko
        self.numer_buta = numer_buta
    @property
    def imie(self):
        return self._imie
    @imie.setter
    def imie(self, value):
        if not isinstance(value, str) or not (3 <= len(value) <= 20):
            raise ValueError("Imię musi być łańcuchem znaków długości od 3 do 20 znaków.")
        self._imie = value
        
    @property
    def nazwisko(self):
        return self._nazwisko
    @nazwisko.setter
    def nazwisko(self, value):
        if not isinstance(value, str) or not (3 <= len(value) <= 20):
            raise ValueError("Nazwisko musi być łańcuchem znaków długości od 3 do 20 znaków.")
        self._nazwisko = value
        
    @property
    def numer_buta(self):
        return self._numer_buta
    @numer_buta.setter
    def numer_buta(self, value):
        if not isinstance(value, int) or not (20 <= value <= 50):
            raise ValueError("Numer buta musi być liczbą całkowitą w zakresie od 20 do 50.")
        self._numer_buta = value

    # tutaj metoda __str__
    def __str__(self):
        return f"{self.imie} {self.nazwisko} ma numer buta {self.numer_buta}."
    
osoba = Osoba("Julia", "Bąk", 36)
print(osoba)
````

#### Zadanie 8.
Zaktualizować klasę `Prostokąt` o odpowiednie mechanizmy zabezpieczeń. W przypadku próby przypisania danych, które nie spełniają założeń program powinien wypisać odpowiedni komunikat na ekran. Boki prostokąta powinny być liczbami całkowitymi lub zmiennoprzecinkowymi o wartości większej niż $0$.

````python
class Prostokat:
    def __init__(self, a, b):
        # tutaj inicjalizacja atrybutów
        self.a = a
        self.b = b

    @property
    def a(self):
        return self._a
    @a.setter
    def a(self, value):
        if not isinstance(value, (int, float)) or value <= 0:
            raise ValueError("Długość boku musi być liczbą dodatnią.")
        self._a = value

    @property
    def b(self):
        return self._b
    @b.setter
    def b(self, value):
        if not isinstance(value, (int, float)) or value <= 0:
            raise ValueError("Długość boku musi być liczbą dodatnią.")
        self._b = value

    # tutaj pozostałe metody
    def obwod(self):
        return 2 * (self.a + self.b)
    
    def pole(self):
        return self.a * self.b
    
prostokat = Prostokat(3,7)
print(prostokat.obwod())
print(prostokat.pole())
````

#### Dziedziczenie
Mechanizm dziedziczenia pozwala na rozszerzanie funkcjonalności istniejącej klasy poprzez odziedziczenie jej atrybutów i metod w klasie która z niej dziedziczy. Dzięki temu w nowo powstałej klasie można się odwoływać do atrybutów i metod, które nie zostały w tej klasie zaimplementowane tylko odziedziczone. Aby dziedziczyć z innej klasy to podczas definiowania klasy po jej nazwie należy w nawiasie podać nazwy klas, z których chcemy dziedziczyć.

```Python
class klasa1:
    def __init__(self, parametr1):
        self.parametr1 = parametr1

    @property
    def parametr1(self):
        return self._parametr1
    @parametr1.setter
    def parametr1(self, wartosc):
        if not isinstance(wartosc, float):
            raise AttributeError('Podana wartość musi być typu zmiennoprzecinkowego')
        self._parametr1 = wartosc

class klasa2(klasa1):
    def __init__(self, parametr1, parametr2):
        super().__init__(parametr1)
        self.parametr2 = parametr2

    @property
    def parametr2(self):
        return self._parametr2
    @parametr2.setter
    def parametr2(self, wartosc):
        if not isinstance(wartosc, bool):
            raise AttributeError('Podana wartość musi być typu logicznego')
        self._parametr2 = wartosc
```

Widać tutaj, że dzięki mechanizmowi dziedziczenia w klasie `klasa2` nie ma już potrzeby implementować ponownie mechanizmów zabezpieczenia atrybutu `parametr1`. Został on już zabezpieczony w klasie `klasa1` i ten cały mechanizm został odziedziczony. Odbyło się to poprzez wywołanie konstruktora klasy macierzystej wywołanej przez słowo kluczowe `super()`.

Jeżeli zachodzi potrzeba mamy możliwość nadpisania ciała wybranej metody. Nazywa się to **przeciążaniem metody**. Polega po prostu na zaimplementowaniu metody jeszcze raz w klasie, która dziedziczy, zgodnie z potrzebą.

#### Zadanie 10.
Napisać klasę Student, która będzie dziedziczyć z klasy `Osoba`. Student posiada dodatkowy atrybut `numer_albumu`, który jest opatrzony mechanizmem bezpieczeństwa dbającym o to aby numer albumu był tekstem składającym się z $6$ cyfr. Należy nadpisać metodą `__str__` w taki sposób aby oprócz imienia i nazwiska student przedstawiał się numerem albumu.

````python
class Student(Osoba):
    def __init__(self, imie, nazwisko, numer_albumu, numer_buta):
        super().__init__(imie, nazwisko, numer_buta)
        
        self.numer_albumu = numer_albumu
        
    @property
    def numer_albumu(self):
        return self._numer_albumu
    @numer_albumu.setter
    def numer_albumu(self, wartosc):
        if not isinstance(wartosc,str):
            raise AttributeError("Numer albumu musi być typu tekstowego.")
        if len(wartosc) != 6:
            raise AttributeError("Numer albumu musi zawierać 6 znaków.")
        self._numer_albumu = wartosc
        
    # tutaj metoda __str__
    def __str__(self):
        return f'({self.imie} {self.nazwisko} ma numer albumu {self.numer_albumu})'
        
student = Student('Jan', 'Kowalski', '123456', 45)
print(student)
````

Wysokość jest liczbą rzeczywistą większą od zera. Należy nadpisać metodę `pole`, aby prawidłowo obliczać pole równoległoboku. 

Podpowiedź: 
$$\text{pole równoległoboku} = \text{długość boku, na który opada wysokość} \cdot \text{wysokość}$$


````python
class Rownoleglobok(Prostokat):
    def __init__(self, a, b, h):
        super().__init__(a, b)
        
        self.h = h
        
    @property
    def h(self):
        return self._h
    @h.setter
    def h(self, wartosc):
        if not isinstance(wartosc, (int, float)):
            raise AttributeError("Wysokość równoległoboku musi być liczbą rzeczywistą")
        elif wartosc <= 0:
            raise AttributeError("Wysokość musi być liczbą dodatnią")
        self._h = wartosc
        
    # tutaj przeciążenie metody pole
    def pole(self):
        return self.a * self.h

    
rownoleglobok = Rownoleglobok(300,200,150)
print(rownoleglobok.obwod())
print(rownoleglobok.pole())
````
