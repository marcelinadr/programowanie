### Zadanie 1. (2 pkt.)
Napisać funkcję NWD, której zadaniem będzie obliczenie największego wspólnego dzielnika dwóch liczb. Pseudokod algorytmu Euklidesa:

```Pseudo
funkcja NWD(a, b)
    dopóki b ≠ 0
        c := a mod b
        a := b
        b := c
    zwróć a
```

````python
def NWD (a,b):
    while b != 0 :
        c = a%b
        a = b
        b = c
    return a
a = int(input("Podaj pierwszą liczbę: "))
b = int(input("Podaj drugą liczbę: "))
print(NWD(a,b))
````

### Zadanie 2. (2 pkt.)
Napisz program, który pobierze od użytkownika liczbę naturalną większą od $3$ i następnie wypisze na ekran romb jak poniżej (tutaj dla $N=4$). Jeżeli otrzymana liczba jest parzysta to należy zwiększyć ją o $1$.

````python
n = int(input("Podaj liczbe:"))
if n%2 == 0:
    n += 1
if n > 3:
    srodek = n//2
    for i in range(n):
        if i <= srodek:
            spacje = " " * (srodek - i)
            gwiazdeczki = "*" * (2 * i + 1)
        else:
            spacje = " " * (i - srodek)
            gwiazdeczki = "*" * (2 * (n - i - 1) + 1)
        print(spacje + gwiazdeczki)
else:
  print("Nie sepełniono warunków. Podaj liczbę większą niż 3")
````

### Zadanie 6. (5 pkt.)
Napisz klasę `Motocykl`. Klasa przyjmuje następujące parametry:
* marka - tekst długości od 3 do 20 znaków
* klasa - tekst o możliwych wartościach: *sport*, *tourer*, *chopper*, *skuter*.
* moc - liczba naturalna o maksymalnej wartości $300$

Klasa implementuje następujące metody:
* jedz() - wypisuje na ekran informację o tym, że motocykl się przemieszcza
* stoj() - wypisuje na ekran informację o tym, że motocykl się zatrzymał

Dodatkowo, przeciąż metodę \_\_str\_\_() aby po wypisaniu obiektu wypisony został napis: *Motocykl {marka} klasy {klasa} o mocy maksymalnej {moc} KM*.

````python
class Motocykl:
  def __init__(self, marka, klasa, moc):
    self.marka = marka
    self.klasa = klasa
    self.moc = moc

  @property
  def marka(self):
    return self._marka

  @marka.setter
  def marka(self, wartosc):
    if 3 <= len(wartosc) <= 20:
      self._marka = wartosc
    else:
        raise ValueError("Marka musi mieć od 3 do 20 znakow")
  @property
  def klasa(self):
    return self._klasa

  @klasa.setter
  def klasa(self, wartosc):
    if wartosc in ["sport", "tourer", "chopper", "skuter"]:
      self._klasa = wartosc
    else:
        raise ValueError ("Nieznana klasa")

  @property
  def moc(self):
    return self._moc
  @moc.setter
  def moc(self, wartosc):
    if wartosc <= 300:
      self._moc = wartosc
    else:
        raise ValueError("Moc musi być mniejsza lub rowna 300")

  def jedz(self):
    print(f"Motocykl {self.marka} przemieszcza się.")

  def stoj(self):
    print(f"Motocykl {self.marka} zatrzymał się.")

  def __str__(self):
    return (f"Motocykl {self.marka} klasy {self.klasa} o mocy maksymalnej {self.moc} KM")


moj_motocykl = Motocykl("BMW", "sport", 200)
moj_motocykl.jedz()
moj_motocykl.stoj()
print(moj_motocykl)
````

### Zadanie 7. (5 pkt.)
Napisz klasę `Punkt`. Jako parametry przyjmuje współrzędne punktu na płaszczyźnie. Posiada dwie metody:
* przesun(dx, dy) - przyjmuje wektor, przesuwa punkt o podany wektor,
* odleglosc(inny) - jako parametr przyjmuje obiekt innego punktu, oblicza odległość bieżącego punktu od podanego w parametrze. Wykorzystaj odległość Bray-Curtis:
$$d_{BC}=\frac{\sum_{i=1}^n|x_i-y_i|}{\sum_{i=1}^n(x_i+y_i)}$$

Aby obliczyć wartość bezwzględną wykorzystaj funkcję `abs()`

Utwórz dwa obiekty tej klasy i oblicz odległość między nimi.

````python
class Punkt:
  def __init__(self, x, y):
    self.x = x
    self.y = y

  def przesun(self, dx, dy):
    self.x += dx
    self.y += dy

  def odleglosc(self, inny):
    licznik = abs(self.x - inny.x) + abs(self.y - inny.y)
    mianownik = abs(self.x + inny.x) + abs(self.y + inny.y)
    if mianownik == 0:
      return 0.0
    else:
      return licznik / mianownik

p1 = Punkt(2, 8)
p2 = Punkt(9, 13)

wynik = p1.odleglosc(p2)

print(f"Punkt 1: ({p1.x}, {p1.y})")
print(f"Punkt 2: ({p2.x}, {p2.y})")
print(f"Odległość Bray-Curtis: {wynik:.4f}")
````

### Zadanie 8. (5 pkt.)
Napisz klasę `Licznik`. Jako parametr przyjmuje element startowy będący liczbą całkowitą - domyślnie $0$. Klasa posiada następujące metody:
* inkrementuj() - zwiększa wartość licznika o 1,
* dekrementuj() - zmniejsza wartość licznika o 1,
* resetuj() - przywraca wartość licznika do wartości początkowej,
* wartosc - metoda obsługiwana jako atrybut, bez możliwości edycji, zwraca aktualną wartość licznika.

````python
class Licznik:
    def __init__(self, start=0):
        self._start = start
        self._aktualna = start

    @property
    def wartosc(self):
        return self._aktualna

    def inkrementuj(self):
        self._aktualna += 1

    def dekrementuj(self):
        self._aktualna -= 1

    def resetuj(self):
        self._aktualna = self._start

l = Licznik(15)   
l.inkrementuj()    
l.inkrementuj()   
l.dekrementuj()    

print(f"Aktualna wartość: {l.wartosc}")

l.resetuj()
print(f"Wartość po resecie: {l.wartosc}")
````

### Zadanie 9. (5 pkt.)
Napisz klasę `Temperatura`. Jako parametr przyjmuje liczbę rzeczywistą nieujemną będącą temperaturą wyrażoną w Kelvinach. Aktualizacja temperatury odbywa się poprzez użycie odpowiedniej metody jako atrybutu. Obiekt klasy przechowuje temperaturę. Klasa implementuje następujące metody:
* fahrenheit - konwertuje aktualną temperaturę na stopnie Fahrneheita i zwraca ją do programu, metoda obsługiwana jak atrybut. Po przypisaniu nowej wartości konwertuje ją na Kelviny i aktualizuje wartość aktualnej temperatury.

Podpowiedź:
$$T_F = (T_K - 273.15) * \frac{9}{5} + 32$$
gdzie:
- $T_F$ - temperatura wyrażona w stopniach Fahrenheita,
- $T_K$ - temperatura wyrażona w stopniach Celsjusza.

Podpowiedź: $0$ Kelvinów oznacza zero absolutne i w rozumieniu fizyki klasycznej temperatura nie może osiągnąć wartości poniżej zera absolutnego.

````python
class Temperatura:
  def __init__(self, kelviny):
    if kelviny < 0:
      raise ValueError("Temperatura nie może być niższa niż 0 Kelvinów!")
    self._kelviny = float(kelviny)

  @property
  def fahrenheit(self):
    return (self._kelviny - 273.15) * 9/5 + 32

  @fahrenheit.setter
  def fahrenheit(self, wartosc_f):
    nowe_kelviny = (wartosc_f - 32) * 5/9 + 273.15

    if nowe_kelviny < 0:
      raise ValueError("Wynikowa temperatura nie może być niższa niż 0 Kelvinów!")

    self._kelviny = nowe_kelviny

t = Temperatura(754)  
print(f"Temperatura w Kelvinach: {t._kelviny} K")
print(f"Temperatura w Fahrenheitach: {t.fahrenheit:.2f} F")


t.fahrenheit = 32
print(f"\nPo ustawieniu 32 F:")
print(f"Temperatura w Kelvinach: {t._kelviny:.2f} K")
````
