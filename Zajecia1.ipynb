W Pythonie operacja odwrócenia bitowego odbywa się zgodnie z wzorem 
$$~x = -(x + 1)$$
Jest to powiązane ze sposobem kodowania liczb całkowitych metodą uzupełnienia do dwóch (U2). Przykładowo:
$$5=101=0b101=00000101_{(U2)}$$
$$~0b101=11111010_{(U2)}$$
$$11111010_{(U2)}=-6=-0b110$$

Dodatkowe informacje: https://pl.wikipedia.org/wiki/Kod_uzupe%C5%82nie%C5%84_do_dw%C3%B3ch

````python
bin(_0b101) # =-0b110
````
